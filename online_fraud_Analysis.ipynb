

<h1 style="font-family: 'times new roman';">Introduction</h1>


<div style="font-family: 'times new roman';">
This project focuses on implementing a Decision Tree classification model to address the significant challenge of online payment fraud. We utilize the "Online Payment Fraud Detection" dataset sourced from Kaggle, which comprises synthetic data generated by the PaySim simulator to mimic real-world mobile money transactions. This extensive dataset includes features like transaction type, amount, and account balances, along with a crucial label indicating whether a transaction is fraudulent (isFraud = 1) or legitimate (isFraud = 0). The primary objective is to train an interpretable Decision Tree model capable of accurately identifying potentially fraudulent activities within this dataset.
